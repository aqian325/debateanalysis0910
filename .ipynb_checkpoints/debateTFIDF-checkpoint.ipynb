```python
import re
import string

def remove_prefixes_and_punctuation(file_path, prefix, output_file):
    # Compile the regex pattern to match the prefix at the start of a line
    pattern = re.compile(rf"^{prefix}: ", re.MULTILINE)
    
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # trying to remove fancy apostrophes
    modified_content = re.sub(r"[‘’`]", "'", content)

    # Remove the prefix from the content
    modified_content = pattern.sub('', content)

    # also remove punctuation and numbers
    punctuation_without_apostrophe = string.punctuation.replace("'", "")
    modified_content = modified_content.translate(str.maketrans('', '', punctuation_without_apostrophe))
    modified_content = modified_content.translate(str.maketrans('', '', string.digits))
    
    # Write the modified content to a new file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(modified_content)

def remove_just_prefixes(file_path, prefix, output_file):
    # Compile the regex pattern to match the prefix at the start of a line
    pattern = re.compile(rf"^{prefix}: ", re.MULTILINE)
    
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    modified_content = re.sub(r"[‘’`]", "'", content)

    # Remove the prefix from the content
    modified_content = pattern.sub('', content)
    
    # Write the modified content to a new file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(modified_content)


# Define file paths and prefixes
harris_file = 'text-documents/Harris_Debate_0910.txt'
trump_file = 'text-documents/Trump_Debate_0910.txt'
harris_output = 'text-documents/Noname_Harris_Debate_0910.txt'
trump_output = 'text-documents/Noname_Trump_Debate_0910.txt'
harris_punctuation = 'text-documents/Yesdot_Noname_Harris_Debate_0910.txt'
trump_punctuation = 'text-documents/Yesdot_Noname_Trump_Debate_0910.txt'

biden_file = 'text-documents/Biden_Debate_Transcript.txt'
first_trump_file = 'text-documents/First_Trump_Debate_Transcript.txt'
biden_output = 'text-documents/Noname_Biden_Debate_0627.txt'
first_trump_output = 'text-documents/Noname_First_Trump_Debate_0627.txt'
biden_punctuation = 'text-documents/Yesdot_Noname_Biden_Debate_0627.txt'
first_trump_punctuation = 'text-documents/Yesdot_Noname_First_Trump_Debate_0627.txt'

# Process each file
# remove_prefixes_and_punctuation(harris_file, "VICE PRESIDENT KAMALA HARRIS", harris_output)
# remove_prefixes_and_punctuation(trump_file, "FORMER PRESIDENT DONALD TRUMP", trump_output)
remove_prefixes_and_punctuation(biden_file, "BIDEN", biden_output)
remove_prefixes_and_punctuation(first_trump_file, "TRUMP", first_trump_output)
remove_just_prefixes(harris_file, "VICE PRESIDENT KAMALA HARRIS", harris_punctuation)
remove_just_prefixes(trump_file, "FORMER PRESIDENT DONALD TRUMP", trump_punctuation)
remove_just_prefixes(biden_file, "BIDEN", biden_punctuation)
remove_just_prefixes(first_trump_file, "TRUMP", first_trump_punctuation)



```

```python
import altair as alt
import numpy as np

# Terms in this list will get a red dot in the visualization
term_list = ['', '']

# adding a little randomness to break ties in term ranking
top_tfidf_plusRand = top_tfidf.copy()
top_tfidf_plusRand['tfidf'] = top_tfidf_plusRand['tfidf'] + np.random.rand(top_tfidf.shape[0])*0.0001

# base for all visualizations, with rank calculation
base = alt.Chart(top_tfidf_plusRand).encode(
    x = 'document:N',
    y = 'rank:O'
).transform_window(
    rank = "rank()",
    sort = [alt.SortField("tfidf", order="descending")],
    groupby = ["document"],
)

# heatmap specification
heatmap = base.mark_rect().encode(
    color = 'tfidf:Q'
)

# red circle over terms in above list
circle = base.mark_circle(size=100).encode(
    color = alt.condition(
        alt.FieldOneOfPredicate(field='term', oneOf=term_list),
        alt.value('red'),
        alt.value('#FFFFFF00')        
    )
)

# text labels, white for darker heatmap colors
text = base.mark_text(baseline='middle').encode(
    text = 'term:N',
    color = alt.condition(alt.datum.tfidf >= 0.23, alt.value('white'), alt.value('black'))
)

# display the three superimposed visualizations
(heatmap + circle + text).properties(width = 600)
```

```python
from collections import Counter
import pandas as pd

def word_count_and_unique(file_paths):
    data = []
    
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            words = content.split()
            total_words = len(words)
            unique_words = len(set(words))
            unique_percentage = (unique_words / total_words) * 100 if total_words > 0 else 0
            
            data.append({
                'File': file_path,
                'Word Count': total_words,
                'Unique Word Count': unique_words,
                'Percentage of Unique Words (%)': unique_percentage
            })
    
    # Convert the list of dictionaries into a pandas DataFrame for a clean table format
    df = pd.DataFrame(data)
    
    # Sort the DataFrame by 'Percentage of Unique Words (%)' in descending order
    df_sorted = df.sort_values(by='Percentage of Unique Words (%)', ascending=False)
    
    return df_sorted

file_paths = [
    'text-documents/Noname_Harris_Debate_0910.txt',
    'text-documents/Noname_Trump_Debate_0910.txt',
    'text-documents/Noname_Biden_Debate_0627.txt',
    'text-documents/Noname_First_Trump_Debate_0627.txt'
]

df_sorted = word_count_and_unique(file_paths)

print(df_sorted)
df_sorted
```
